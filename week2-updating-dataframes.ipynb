```python
# About the data

# well use palmer penguins

# Read in data
URL = 'https://raw.githubusercontent.com/allisonhorst/palmerpenguins/main/inst/extdata/penguins.csv'
penguins = pd.read_csv(URL)

penguins.head()
```

```python
penguins.info()
```

```python
penguins.describe()
```

```python
penguins[['species','island','sex','year']].nunique()
```

```python
penguins[['bill_length_mm', 'bill_depth_mm']].plot(kind = 'box',
                                                  title = "Bill Length & Bill Depth Comparison",
                                                  )
```

```python
df_fem_gen = penguins[(penguins['sex'] == "female") & (penguins['species'] == 'Gentoo')]

df_fem_gen['flipper_length_mm'].plot(kind = 'hist',
                                    title = 'frew of flipper length in female gentoo penguins',
                                    xlabel = 'Flipper length mm',
                                    color = 'hotpink')
```

```python
#Creating a one liner for the graph above without subsetting

penguins.loc[(penguins['sex'] == "female") & (penguins['species'] == 'Gentoo'), 'flipper_length_mm'].plot(kind='hist')
```
